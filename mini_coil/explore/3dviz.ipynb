```python
import numpy as np
import os

DATA_DIR = os.path.join('..', '..', 'data')
```

```python
embeddings = np.load(os.path.join(DATA_DIR, 'parametric-umap-8000-jina-small-4-augmented', 'embeddings-bat.npy'))
```

```python
import plotly.express as px

# Create 3D scatter plot
fig = px.scatter_3d(
    x=embeddings[:, 0],
    y=embeddings[:, 1],
    z=embeddings[:, 2],
    opacity=0.5,
    title='3D Embeddings Visualization'
)

# Update the layout for better visualization
fig.update_traces(marker=dict(size=3))
fig.update_layout(
    scene = dict(
        xaxis_title='X',
        yaxis_title='Y', 
        zaxis_title='Z',
        aspectmode='cube'
    )
)

fig.show()
```
