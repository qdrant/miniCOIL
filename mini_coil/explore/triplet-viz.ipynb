```python
import numpy as np
import os


DATA_DIR = os.path.join('..', '..', 'data')
```

```python
embeddings_2d = np.load(os.path.join(
    DATA_DIR,
    'triplets-8000-mxbai-large-jina-small-2-augmented',
    'word-models',
    'validation',
    'bat-validation.npy'
))
```

```python
# Create scatter plot of 2D embeddings
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.scatter(embeddings_2d[:,0], embeddings_2d[:,1], alpha=0.5, s=1)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('2D Embeddings')
plt.show()
```

```python
# Convert 3D coordinates to latitude/longitude
import matplotlib.pyplot as plt

# Calculate latitude and longitude from 3D coordinates
latitude = np.arcsin(embeddings_3d[:, 2]) # z coordinate determines latitude 
longitude = np.arctan2(embeddings_3d[:, 1], embeddings_3d[:, 0]) # y/x determines longitude

# Plot the points on a 2D lat/long grid
plt.figure()
plt.scatter(longitude, latitude, alpha=0.5, s=1)
plt.xlabel('Longitude (radians)')
plt.ylabel('Latitude (radians)') 
plt.grid(True)
plt.show()
```
